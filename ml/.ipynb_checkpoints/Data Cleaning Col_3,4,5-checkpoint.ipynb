# Column 5 : Degree Percentage

```python
df.head()
```

```python
df.rename(columns = {"Degree Percentage (Degree Marks Obtained / Total Marks)": "Degree Percentage"}, inplace=True)
```

```python
df.head()
```

```python
df['Degree Percentage'].value_counts()
```

```python
df['Degree Percentage'].unique()
```

```python
df['Degree Percentage'] = df['Degree Percentage'].astype(str).str.rstrip('%')
df['Degree Percentage'].unique()
```

```python
df.isnull().sum()
```

```python
df['Degree Percentage'].shape
```

```python
df.shape
```

```python
df['Degree Percentage'].unique()
```

```python
def remove():
    df['Degree Percentage'].replace(['na', 'nan', '-','Na', '0','NIL','None', '.','Not Applicable', 'N.A.'
                                    , 'none','NL','Not applicable','----',"I don't have updated S.E results"
                                    , 'No', 'N.A'], None, inplace=True)
    df['Degree Percentage'].replace('7.75 CGPA APPROX. (60-70)', "65", inplace=True)
remove()
```

```python
df['Degree Percentage'].unique()
```

```python
df['Degree Percentage'].isnull().sum()
```

```python
df['Degree Percentage'] = df['Degree Percentage'].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Degree Percentage'] = df['Degree Percentage'].fillna(df['Degree Percentage'].mean())
df.isnull().sum()
```

```python
df1 = df
df1.head()
```

```python
df = df[df['Degree Percentage'] > 10]
```

```python
df.describe()
```

```python
df['Degree Percentage'] =df['Degree Percentage'].round(0)
df.head()
```

```python
df['Degree Percentage'] =df['Degree Percentage'].astype(int)
df.info()
```

```python
df.head()
```
